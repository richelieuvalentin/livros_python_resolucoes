## 2.2 Escreva um algoritmo que leia 3 números inteiros e diferentes e mostre-os em ordem decrescente. Use seleção encadeada.

```python
a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))
c = int(input("Digite o valor de c: "))

if a == b or b == c:
    print("números iguais")
else:
    if a < b and b < c:
        print(f"c: {c}, b: {b}, a: {a}")
    else:
        if b < a and a < c:
            print(f"c: {c}, a: {a}, b: {b}")
        else:
            if a < c and c < b:
                print(f"b: {b}, c: {c}, a: {a}")
            else:
                if c < a and a < b:
                    print(f"b: {b}, a: {a}, c: {c}")
                else:
                    if c < b and b < a:
                        print(f"a: {a}, b: {b}, c: {c}")
                    else:
                        print(f"a: {a}, c: {c}, b: {b}")                       
```

## 2.3 Escreva um algoritmo que calcule as raízes de uma equação do 2ª grau, na forma $ax^2 + bx + c$ levando em consideração a existência de raízes reais


```python
# valores de a, b e c
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))

# Discriminante delta
delta = b ** 2 - (4 * a * c)

if delta > 0:
    x1 = (-b + delta ** (1/2)) / (2 * a)
    x2 = (-b - delta ** (1/2)) / (2 * a)
    print("Delta > 0")
    print(f"As raízes da equação são {x1} e {x2}")
elif delta == 0:
    x = -b / (2 * a)
    print("Delta = 0")
    print(f"A raiz da equação é {x}")
else:
    print("A equação não possui raízes reais!")
```

## 2.4 Tendo como dados de entrada a altura e sexo de uma pessoa, construa um algoritmo que calcule o seu peso ideal, utilizando a seguinte fórmula:
Para homens: (72,7 * h) - 58;
Para mulheres: (62,1 * h) - 44,7.

```python
sexo = input("Qual o seu sexo? Digite F para feminino e M para masculino: ")
altura = float(input("Qual a sua altura? Digite o valor em metros: "))

if sexo == "F":
    peso_ideal = (62.1 * altura) - 44.7
    print(f"Seu peso ideal é {peso_ideal:.2f}")
elif sexo == "M":
    peso_ideal = (72.7 * altura) - 58
    print(f"Seu peso ideal é {peso_ideal:.2f} kg")
else:
    print("Erro! Digite um valor válido!")
```

## 2.8 Elabore um algoritmo que calcule o que deve ser pago por um produto, considerando o preço normal de etiqueta e a escolha da condição de pagamento. Utilize os códigos da tabela a seguir para ler qual a condição de pagamento escolhida e efetuar o cálculo adequado.
| Código | Condição de pagamento                                     |
| -----: | --------------------------------------------------------- |
|      1 | À vista em dinheiro ou cheque, recebe 10% de desconto     |
|      2 | À vista no cartão de crédito, recebe 5% de desconto       |
|      3 | Em duas vezes, preço normal de etiqueta sem juros         |
|      4 | Em três vezes, preço normal de etiqueta mais juros de 10% |


```python
codigo = int(input("Digite o código: "))
preco_etiqueta = float(input("Digite o preço de etiqueta: "))

match codigo:
    case 1:
        valor_a_pagar = preco_etiqueta - (preco_etiqueta * (10 / 100))
        print(f"O valor a pagar é R$ {valor_a_pagar:.2f}")
    case 2:
        valor_a_pagar = preco_etiqueta - (preco_etiqueta * (5 / 100))
        print(f"O valor a pagar é R$ {valor_a_pagar:.2f}")
    case 3:
        valor_a_pagar = preco_etiqueta
        print(f"O valor a pagar é R$ {valor_a_pagar:.2f}")
    case 4:
        valor_a_pagar = preco_etiqueta + (preco_etiqueta * (10 / 100))
        print(f"O valor a pagar é R$ {valor_a_pagar:.2f}")
```

## 2.9 Elabore um algoritmo que leia o valor de dois números inteiros e a operação aritmética desejada; calcule, então, a resposta adequada. Utilize os símbolos da tabela a seguir para ler qual a operação aritmética escolhida.
| Símbolo | Operação aritmética |
| ------- | ------------------- |
| +       | Adição              |
| -       | Subtração           |
| *       | Multiplicação       |
| /       | Divisão             |


```python
numero1 = int(input("Digite o primeiro inteiro: "))
numero2 = int(input("Digite o segundo número inteiro: "))
operacao = input("Digite a operacao '-', '+', '*' ou '/': ")

match operacao:
    case "-":
        print(numero1 - numero2)
    case "+":
        print(numero1 + numero2)
    case "*":
        print(numero1 * numero2)
    case "/":
        if numero2 == 0:
            print("Erro: divisão por zero não é permitida.")
        else:
            print(numero1 / numero2)
    case _:
        print("Operação inválida! Use apenas '-', '+', '*' ou '/'.")
```

## 2.10 O IMC – Índice de Massa Corporal – é um critério da Organização Mundial de Saúde para indicar a condição de peso de uma pessoa adulta. A fórmula é IMC = peso / (altura)². Elabore um algoritmo que leia o peso e a altura de um adulto e mostre sua condição.
| IMC em adultos  | Condição       |
| --------------- | -------------- |
| abaixo de 18,5  | abaixo do peso |
| entre 18,5 e 25 | peso normal    |
| entre 25 e 30   | acima do peso  |
| acima de 30     | obeso          |


```python
altura = float(input("Digite sua altura: "))
peso = float(input("Digite seu pesoa"))

imc = peso / (altura ** 2)

if imc < 18.5:
    print("abaixo do peso")
elif 18.5 <= imc <= 25:
    print("peso normal")
elif 25 < imc <= 30:
    print("Acima do peso")
elif imc > 30:
    print("obeso")
else:
    print("Inválido")
```
